```python
import pandas as pd

# Sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 35, 61, 45, 70]
}

df = pd.DataFrame(data)
```

```python
def categorize_age(age):
    if age <= 30:
        return 'Young'
    elif age <= 60:
        return 'Middle-aged'
    else:
        return 'Senior'
```

```python
df['Age_Group'] = df['Age'].apply(categorize_age)

print(df)
```

```python
import matplotlib.pyplot as plt

# Sample data
departments = ['HR', 'Engineering', 'Sales', 'Marketing', 'Customer Support']
employee_counts = [12, 45, 28, 18, 22]

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(departments, employee_counts, color='skyblue')

# Add labels and title
plt.xlabel('Number of Employees')
plt.title('Employee Count by Department')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Optional: Show values on the bars
for index, value in enumerate(employee_counts):
    plt.text(value + 0.5, index, str(value), va='center')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data points
x1 = [1, 2, 3]
y1 = [4, 5, 6]

x2 = [4, 5, 6]
y2 = [1, 2, 3]

# Plot each group separately
plt.scatter(x1, y1, color='blue', label='Group A')
plt.scatter(x2, y2, color='green', label='Group B')

plt.title("Categorized Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
